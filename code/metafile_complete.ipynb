```python
for i in range(np.shape(meta)[0]):
    name = '*' + meta.loc[i,'Image_id']+'.nii'
    filename = fnmatch.filter(os.listdir('/csc/epitkane/data/ADNI/AD_DL_16_04_22/'), name)
    if len(filename)> 0:
        meta.loc[i,'PATH'] = '/csc/epitkane/data/ADNI/AD_DL_16_04_22/' + filename[0]
```
